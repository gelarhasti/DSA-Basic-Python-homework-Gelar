```python
googleps.drop_duplicates(subset='App', inplace=True)
```

```python
googleps['Installs'] = googleps['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
```

```python
googleps['Installs'] = googleps['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)
```

```python
googleps['Installs'] = googleps['Installs'].astype(int)
```

```python
## Match Apps and Category based on provided dataset! 
array = ['FIFA Soccer', 'Dropbox', 'Subway Surfers', 'DU Battery Saver - Battery Charger & Battery Life', 'Pinterest', 'Flipboard: News For Our Time','File Commander - File Manager/Explorer','WhatsApp Messenger']
googleps.loc[googleps['App'].isin(array)]
```

```python
## The following applications are installed by more than 1 billion users, EXCEPT ...

googleps.loc[googleps['Installs'] >= 1000000000]
```

```python
## Arrange top 5 COMMUNICATION category apps from the highest number of reviews to the lowest! 

catapp = googleps[googleps['Category']=='COMMUNICATION']
catapp.sort_values(by=['Reviews'], ascending=False).head(7)
```

```python
## Arrange top 3 game category apps based on number of installed and rating!

catgame = googleps[googleps['Category']=='GAME']
catgame.sort_values(by=['Installs','Rating'], ascending=False).head(3)
```
